```python
from happytransformer import HappyGeneration, GENTrainArgs, GENSettings
from torch import manual_seed
from random import randint
```

```python
happy_gen = HappyGeneration(model_type="GPT2", model_name="penguinwriter/artificer_intelligence")
```

#Usage

This model generates text based on the following format based on a given prompt:

keyword1 (keyword2 ... keyword6)::Name,Type,[rarity],Attunement Y/N,Attunement Requirements,[Description]

To generate a magic item, place your keywords you want in your description, seperated by spaces, before the double colon (::).

If you want to give your magic item a name, type, etc., place it after the double colon.

An example of a prompt is shown

time manipulation katana::Timekeeper,Weapon (Longsword),['legendary'],Y,nan,['When you hit with a melee attack using this sword, you can choose to trap the target in a temporal loop.

#Note that generation is not perfect 100% of the time. You may need to "wrestle" with your prompt to get it working



```python
#Random Seed - if you want your output to be deterministic, comment out this line
manual_seed(randint(0, 9999999999))

#Arguments are explained more @ https://happytransformer.com/text-generation/settings/
args = GENSettings( min_length=250, max_length=600, do_sample=True,early_stopping=True,temperature=0.8,)
prompt = """::Spear of the Jungle Dragon,"""
result = happy_gen.generate_text(prompt, args=args)


print(prompt+result.text)
```
